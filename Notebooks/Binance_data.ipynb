```python
import ccxt
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from binance.client import Client
import time
from datetime import datetime, timedelta

INVERSION_INICIAL = 4000.0  # en euros
# -----------------------------
# EMAIL CONFIG
# -----------------------------
EMAIL_FROM     = "jlahoz.ferrandez@gmail.com"
EMAIL_PASSWORD = "tvli cxgk duwh yzdd"
EMAIL_TO       = "ines_farrera@hotmail.com"

def send_email(subject, body):
    msg = MIMEMultipart()
    msg['From'] = EMAIL_FROM
    msg['To'] = EMAIL_TO
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))
    
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(EMAIL_FROM, EMAIL_PASSWORD)
        server.send_message(msg)
        server.quit()
        print(f"üìß Email sent")
    except Exception as e:
        print(f"‚ö†Ô∏è Error sending email: {e}")

# -----------------------------
# BINANCE CONNECTION
# -----------------------------
api_key    = "9dinTOw9F9x1CXg4qCWhjk7Ug7QgHlMC8jRMDI7HKoiBX41HnC35q9aT16Qz06O7"
api_secret = "qOkElCBDPvnWlLsqDa3FHgk1LeBi8TzFUa2tj5VAOOFqTOeUxAojRrL5MDXsv189"
client     = Client(api_key, api_secret)

# -----------------------------
# GET BALANCES
# -----------------------------
def get_balances():
    return client.get_account()['balances']

# -----------------------------
# FETCH EUR PRICE FOR EACH ASSET
# -----------------------------
def get_eur_price(symbol):
    """Obtiene precio de symbol en EUR"""
    try:
        ticker = client.get_symbol_ticker(symbol=f"{symbol}EUR")
        return float(ticker['price'])
    except:
        try:
            ticker_usdt = client.get_symbol_ticker(symbol=f"{symbol}USDT")
            ticker_usdt_eur = client.get_symbol_ticker(symbol="EURUSDT")
            return float(ticker_usdt['price']) / float(ticker_usdt_eur['price'])
        except:
            return 0.0

# -----------------------------
# CALCULATE BALANCE IN EUR
# -----------------------------
def calculate_balances_eur():
    balances = get_balances()
    total_eur = 0
    lines = []

    for b in balances:
        symbol = b['asset']
        amount = float(b['free']) + float(b['locked'])
        if amount > 0 and symbol != 'EUR':
            price_eur = get_eur_price(symbol)
            value_eur = amount * price_eur
            total_eur += value_eur
            lines.append(f"{symbol}: {amount:.6f} ‚âà {value_eur:.2f} EUR")

    delta      = total_eur - INVERSION_INICIAL
    porcentaje = (delta / INVERSION_INICIAL) * 100
    
    lines.append(f"\nüí∂ TOTAL ‚âà {total_eur:.2f} EUR")
    lines.append(f"üìà Œî = {delta:+.2f} EUR ({porcentaje:+.2f}%)")
    
    return "\n".join(lines)

# -----------------------------
# WAIT UNTIL NEXT RUN
# -----------------------------
def wait_until(hour=18, minute=0):
    now = datetime.now()
    target = now.replace(hour=hour, minute=minute, second=0, microsecond=0)
    
    if now >= target:
        target += timedelta(days=1)
        
    seconds_to_wait = (target - now).total_seconds()
    print(f"‚è≥ Esperando {int(seconds_to_wait)} segundos hasta las {hour:02d}:{minute:02d}...")
    time.sleep(seconds_to_wait)

# -----------------------------
# MAIN LOOP
# -----------------------------
while True:
    
    wait_until(18, 0)  # Espera hasta las 18:00
    print(f"\nüìÖ Ejecutando script a las {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    balance_report = calculate_balances_eur()
    send_email("üå∏‚ù§Ô∏è Eres una mujer maravillosa üå∏‚ù§Ô∏è", balance_report)
    print(balance_report)
```
