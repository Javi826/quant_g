```python
import requests

url = "https://api.bitget.com/api/v2/mix/market/contracts"
params = {"productType": "USDT-FUTURES"}

res = requests.get(url, params=params)
data = res.json().get("data", [])

# Filtrar contratos activos
contratos_activos = [c for c in data if c.get("symbolStatus") == "normal"]

# Extraer los símbolos
simbolos_activos = sorted([c["symbol"] for c in contratos_activos])
print(simbolos_activos)
print("Número de símbolos activos:", len(simbolos_activos))
```
